```python
# Calculate correlation matrix
correlation_matrix = df[numerical_cols].corr()

print("=== CORRELATION MATRIX ===")
print(correlation_matrix)

# Create correlation heatmap
plt.figure(figsize=(10, 8))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix,
            annot=True,
            cmap='coolwarm',
            center=0,
            square=True,
            fmt='.3f',
            cbar_kws={'shrink': 0.8},
            mask=mask)
plt.title('Feature Correlation Heatmap', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# Identify highly correlated features
print("\n=== HIGHLY CORRELATED FEATURE PAIRS ===")
high_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_val = correlation_matrix.iloc[i, j]
        if abs(corr_val) > 0.7:  # Threshold for high correlation
            high_corr_pairs.append((correlation_matrix.columns[i],
                                  correlation_matrix.columns[j],
                                  corr_val))

if high_corr_pairs:
    for pair in high_corr_pairs:
        print(f"{pair[0]} - {pair[1]}: {pair[2]:.3f}")
else:
    print("No highly correlated feature pairs found (threshold: 0.7)")
```

```python
# Create detailed scatter plots for the most interesting feature pairs
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.ravel()

feature_pairs = [
    (numerical_cols[0], numerical_cols[1]),
    (numerical_cols[0], numerical_cols[2]),
    (numerical_cols[0], numerical_cols[3]),
    (numerical_cols[1], numerical_cols[2]),
    (numerical_cols[1], numerical_cols[3]),
    (numerical_cols[2], numerical_cols[3])
]

colors = ['red', 'blue', 'green']
species_list = df[target_col].unique()

for i, (x_col, y_col) in enumerate(feature_pairs):
    for j, species in enumerate(species_list):
        species_data = df[df[target_col] == species]
        axes[i].scatter(species_data[x_col], species_data[y_col],
                       c=colors[j], label=species, alpha=0.7, s=50)

    axes[i].set_xlabel(x_col, fontsize=10)
    axes[i].set_ylabel(y_col, fontsize=10)
    axes[i].set_title(f'{x_col} vs {y_col}', fontsize=12, fontweight='bold')
    axes[i].legend()
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Perform statistical tests
print("=== STATISTICAL TESTS ===")

# Test for normality (Shapiro-Wilk test)
print("\n1. NORMALITY TESTS (Shapiro-Wilk)")
print("H0: Data is normally distributed")
print("H1: Data is not normally distributed")
print("p-value < 0.05: Reject H0 (not normal)")

for col in numerical_cols:
    stat, p_value = stats.shapiro(df[col])
    print(f"\n{col}:")
    print(f"  Statistic: {stat:.6f}")
    print(f"  P-value: {p_value:.6f}")
    if p_value < 0.05:
        print(f"  Result: Not normally distributed ‚ùå")
    else:
        print(f"  Result: Normally distributed ‚úÖ")

# ANOVA test for each feature across species
print("\n\n2. ANOVA TESTS (One-way)")
print("H0: All species have the same mean for the feature")
print("H1: At least one species has a different mean")
print("p-value < 0.05: Reject H0 (significant difference)")

for col in numerical_cols:
    groups = [df[df[target_col] == species][col] for species in df[target_col].unique()]
    f_stat, p_value = stats.f_oneway(*groups)
    print(f"\n{col}:")
    print(f"  F-statistic: {f_stat:.6f}")
    print(f"  P-value: {p_value:.6f}")
    if p_value < 0.05:
        print(f"  Result: Significant difference between species ‚úÖ")
    else:
        print(f"  Result: No significant difference ‚ùå")
```

```python
print("=" * 60)
print("                  DATA QUALITY REPORT")
print("=" * 60)

# Dataset overview
print(f"\nüìä DATASET OVERVIEW:")
print(f"   ‚Ä¢ Total samples: {len(df)}")
print(f"   ‚Ä¢ Total features: {len(df.columns)}")
print(f"   ‚Ä¢ Numerical features: {len(numerical_cols)}")
print(f"   ‚Ä¢ Target classes: {df[target_col].nunique()}")

# Missing values
missing_count = df.isnull().sum().sum()
print(f"\nüîç DATA COMPLETENESS:")
if missing_count == 0:
    print(f"   ‚Ä¢ Missing values: None ‚úÖ")
    print(f"   ‚Ä¢ Data completeness: 100% ‚úÖ")
else:
    print(f"   ‚Ä¢ Missing values: {missing_count} ‚ùå")
    print(f"   ‚Ä¢ Data completeness: {((len(df) * len(df.columns) - missing_count) / (len(df) * len(df.columns))) * 100:.1f}%")

# Class balance
print(f"\n‚öñÔ∏è CLASS BALANCE:")
class_counts = df[target_col].value_counts()
is_balanced = class_counts.std() / class_counts.mean() < 0.1
if is_balanced:
    print(f"   ‚Ä¢ Dataset is well balanced ‚úÖ")
else:
    print(f"   ‚Ä¢ Dataset may be imbalanced ‚ö†Ô∏è")

for species, count in class_counts.items():
    percentage = (count / len(df)) * 100
    print(f"   ‚Ä¢ {species}: {count} samples ({percentage:.1f}%)")

# Outliers summary
print(f"\nüìà OUTLIER ANALYSIS:")
total_outliers = 0
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = len(df[(df[col] < lower_bound) | (df[col] > upper_bound)])
    total_outliers += outliers
    if outliers > 0:
        print(f"   ‚Ä¢ {col}: {outliers} outliers")

if total_outliers == 0:
    print(f"   ‚Ä¢ No significant outliers detected ‚úÖ")

# Correlation insights
print(f"\nüîó FEATURE RELATIONSHIPS:")
high_corr_count = 0
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > 0.7:
            high_corr_count += 1

if high_corr_count > 0:
    print(f"   ‚Ä¢ {high_corr_count} highly correlated feature pairs found")
    print(f"   ‚Ä¢ Consider feature selection for ML models ‚ö†Ô∏è")
else:
    print(f"   ‚Ä¢ No extremely high correlations detected ‚úÖ")

# ML Readiness Assessment
print(f"\nü§ñ MACHINE LEARNING READINESS:")
ml_score = 0

# Check completeness
if missing_count == 0:
    ml_score += 25
    print(f"   ‚úÖ No missing values")
else:
    print(f"   ‚ùå Missing values need handling")

# Check balance
if is_balanced:
    ml_score += 25
    print(f"   ‚úÖ Well-balanced classes")
else:
    print(f"   ‚ö†Ô∏è Class imbalance may need addressing")

# Check data types
if len(numerical_cols) == len(df.columns) - 1:  # All except target are numerical
    ml_score += 25
    print(f"   ‚úÖ All features are numerical")
else:
    print(f"   ‚ö†Ô∏è Some non-numerical features may need encoding")

# Check size
if len(df) > 100:
    ml_score += 25
    print(f"   ‚úÖ Sufficient sample size")
else:
    print(f"   ‚ö†Ô∏è Small sample size")

print(f"\nüìã OVERALL ML READINESS SCORE: {ml_score}/100")

if ml_score >= 80:
    print("üéâ Dataset is READY for machine learning!")
elif ml_score >= 60:
    print("üëç Dataset is MOSTLY READY with minor preprocessing needed")
else:
    print("‚ö†Ô∏è Dataset needs SIGNIFICANT preprocessing before ML")

print("\n" + "=" * 60)
print("                    END OF ANALYSIS")
print("=" * 60)
```
